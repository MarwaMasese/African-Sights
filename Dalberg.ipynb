```python
import pandas as pd
```

```python
dataset = pd.read_excel("/content/ehealth-kenya-facilities-download-21102015.xls")
```

```python
dataset.head()
```

```python
new_dataset = dataset.query('County == "Nairobi" ')
new_dataset.head()
```

```python
new_dataset.to_csv(r"/content/nairobi.xls", index=False)
```

```python
# Determining the number of health institutions in each district within Nairobi county
counts = new_dataset['District'].value_counts()
print(counts)
```

```python
import matplotlib.pyplot as plt

# Plotting the bar graph
plt.figure(figsize=(10, 6))  # Adjust the figure size if needed
counts.plot(kind='bar')
plt.title('Frequency of Districts')
plt.xlabel('District')
plt.ylabel('Frequency')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability if needed
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Add grid lines for better visualization
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

```python
# Determining the number of health institutions in each Division within Nairobi county
counts = new_dataset['Division'].value_counts()
print(counts)
```

```python
import matplotlib.pyplot as plt

# Plotting the bar graph
plt.figure(figsize=(20, 6))  # Adjust the figure size if needed
counts.plot(kind='bar')
plt.title('Frequency of Health Institutions in Divisions')
plt.xlabel('Division')
plt.ylabel('Frequency')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability if needed
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Add grid lines for better visualization
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```
